```python
# import library 
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
import pandas as pd
```

```python
data = pd.read_excel("dataset-siswa.xlsx")
data
```

```python
x = np.asarray (data)
x
```

```python
z = linkage(x, metric='euclidean')
z
```

```python
# plot dendogram
plt.figure(figsize=(20,10))
plt.title('Dendogram Hierarchical Clustering')
plt.xlabel('data')
plt.ylabel('jarak')
dendrogram(z)
plt.show()
```
